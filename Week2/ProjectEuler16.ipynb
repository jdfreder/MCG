Power digit sum
================
Problem 16
-----------------

2$^{15}$ = 32768 and the sum of its digits is 3 + 2 + 7 + 6 + 8 = 26.

What is the sum of the digits of the number 2$^{1000}$?

```python
pwr=2**15
stpwr=str(pwr)
sumpwr=0
for c in stpwr:
    sumpwr += int(c)
print sumpwr
```

```python
pwr=2**1000
stpwr=str(pwr)
sumpwr=0
for c in stpwr:
    sumpwr += int(c)
print sumpwr
```
